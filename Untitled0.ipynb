```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
import matplotlib.pyplot as plt
 
# Load the solar power data
data = pd.read_csv('solar_power_data.csv')
 
# Remove commas from the 'power' column
data['power'] = data['power'].str.replace(',', '')
 
# Convert 'power' column to float
data['power'] = data['power'].astype(float)
 
# Reshape the data
data = data['power'].values.reshape(-1, 1)
 
# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
data = scaler.fit_transform(data)
seq_length = 1
 
# Split the data into training and test sets
train_size = int(len(data) * 0.8)
test_size = len(data) - train_size
X_train, y_train = data[0:train_size, :], data[seq_length:train_size + seq_length, :]


 
# Convert the data into sequences
def create_sequences(data, seq_length):
   X = []
   y = []
   for i in range(len(data) - seq_length):
       X.append(data[i:i+seq_length])
       y.append(data[i+seq_length])
   return np.array(X), np.array(y)
 
# Create LSTM model
model = Sequential()
model.add(LSTM(units=150, return_sequences=True, input_shape=(seq_length, 1)))
model.add(LSTM(units=50))
model.add(Dense(units=1))
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=4)
# Make predictions
train_predictions = model.predict(X_train)
# Inverse transform the data
train_predictions = scaler.inverse_transform(train_predictions)
y_train = scaler.inverse_transform(y_train)
# Calculate accuracy metric
train_rmse = np.sqrt(np.mean((train_predictions - y_train) ** 2))
# Plot actual and predicted values
plt.figure(figsize=(10, 6))
plt.plot(y_train.flatten(), label='Actual')
plt.plot(train_predictions.flatten(), label='Predicted')
plt.xlabel('Time')
plt.ylabel('Power')
plt.title('Solar Power Prediction')
plt.legend()
plt.show()
# Print accuracy metric
print("Train RMSE:", train_rmse)
```
