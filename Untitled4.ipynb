```python
from scipy.optimize import minimize

import numpy as np

import matplotlib.pyplot as plt

 

# Objective function

def objective(x):

   cost_a = 10

   cost_b = 8

   return cost_a * x[0] + cost_b * x[1]

 

# Constraint function

def constraint(x):

   return x[0] + x[1] - demand

 

# Bounds and initial guess

bounds = ((0, None), (0, None))

x0 = [50, 50]

 

# Define the demand

demand = 100

 

# Solve the optimization problem

solution = minimize(objective, x0, method='SLSQP', bounds=bounds, constraints={'type': 'eq', 'fun': constraint})

# SLSQP -Sequential Least SQuares Programming optimizer.

 

# Optimal power generation values

power_a_opt = solution.x[0]

power_b_opt = solution.x[1]

 

# Plotting the objective function and constraint

power_a = np.linspace(0, 100, 100)

power_b = np.linspace(0, 100, 100)

X, Y = np.meshgrid(power_a, power_b)

Z = objective([X, Y])

constraint_line = demand - X

 

# Plotting the contour plot

fig, ax = plt.subplots()

ax.contourf(X, Y, Z, levels=20, cmap='viridis')

ax.contour(X, Y, constraint_line, levels=[0], colors='red', linewidths=2)

ax.set_xlabel('Power A')

ax.set_ylabel('Power B')

ax.set_title('Constrained Optimization')

ax.scatter(power_a_opt, power_b_opt, color='red', label='Optimal Solution')

ax.legend()

plt.show()

 

# Print the optimal solution

print('Optimal Power Generation:')

print('Plant A:', power_a_opt)

print('Plant B:', power_b_opt)

print('Total Cost:', solution.fun)
```
