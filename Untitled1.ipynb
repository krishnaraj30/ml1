```python

   import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
 
# Load the solar power data
data = pd.read_csv('solar_power_data.csv')
 
# Remove commas from the 'power' column
data['power'] = data['power'].str.replace(',', '')
 
# Convert 'power' column to float
data['power'] = data['power'].astype(float)
 
# Set the date column as the index
data['Day'] = pd.to_datetime(data['Day'])
data.set_index('Day', inplace=True)
 
# Resample the data to daily frequency
data = data.resample('D').mean()
 
# Handling missing values
data = data.fillna(method='ffill')
 
# Split the data into training and test sets
train_size = int(len(data) * 0.8)
train_data, test_data = data[:train_size], data[train_size:]
 
# Fit ARIMA model
model = ARIMA(train_data, order=(10, 0, 150))
model_fit = model.fit()
 
# Plot PACF
fig, ax = plt.subplots(figsize=(10, 4))
plot_pacf(train_data, lags=50, ax=ax)
ax.set_xlabel('Lag')
ax.set_ylabel('Partial Autocorrelation')
ax.set_title('Partial Autocorrelation Function (PACF)')
plt.show()
 
# Plot ACF
fig, ax = plt.subplots(figsize=(10, 4))
plot_acf(train_data, lags=50, ax=ax)
ax.set_xlabel('Lag')
ax.set_ylabel('Autocorrelation')
ax.set_title('Autocorrelation Function (ACF)')
plt.show()
 
# Forecast
forecast = model_fit.forecast(steps=len(test_data))
 
# Calculate accuracy metrics
mse = mean_squared_error(test_data, forecast)
rmse = np.sqrt(mse)
mae = mean_absolute_error(test_data, forecast)
 
# Plot actual and predicted values
plt.figure(figsize=(10, 6))
plt.plot(train_data, label='Training Data')
plt.plot(test_data, label='Actual')
plt.plot(test_data.index, forecast, label='Predicted')
plt.xlabel('Date')
plt.ylabel('Power')
plt.title('Solar Power Prediction using ARIMA')
plt.legend()
```
