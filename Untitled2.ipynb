```python
import numpy as np

import matplotlib.pyplot as plt

# Generate a sparse time series signal

signal_length = 1000

sparse_signal = np.zeros(signal_length)

sparse_signal[[100, 200, 300, 400, 500, 600, 700, 800]] = [2, 4, -3, 1, 5, -2, 3, -4]

# Define the measurement matrix

measurement_matrix = np.random.randn(900, signal_length)

# Take random measurements of the sparse signal

measurements = np.dot(measurement_matrix, sparse_signal)

# Perform sparse signal recovery using L1-norm minimization

recovered_signal = np.linalg.lstsq(measurement_matrix, measurements, rcond=None)[0]

# Plot the original and recovered signals

plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)

plt.plot(sparse_signal)

plt.title('Original Sparse Signal')

plt.subplot(2, 1, 2)

plt.plot(recovered_signal)

plt.title('Recovered Signal')

plt.tight_layout()

plt.show()
```
